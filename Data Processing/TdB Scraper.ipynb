```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import csv
from IPython import display
import re
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.DataFrame(pd.read_csv('ml-latest-small/links.csv'), columns = ["movieId","imdbId","tmdbId"]) 
```

```python
len(df)
```

```python
dicts = {}
keys = range(len(df))
for i in keys:
        dicts[i] = "https://www.themoviedb.org/movie/" + str(df.loc[i, "tmdbId"])
```

```python
with open("tmdbinfofnl.txt","w") as f:
    keys = range(len(df))
    for i in keys:
        resp = requests.get(dicts[i])
        if resp.status_code == 200:
            soup = BeautifulSoup(resp.text)
            overview = soup.find("div", {"class":"overview"})
            youtube = soup.find('a', class_ = 'no_click play_trailer')
            picture = soup.find('a', class_ = 'no_click progressive replace')
            score = soup.find('div', class_ = 'user_score_chart')
            if youtube == None:
                y = ""
            else: 
                y = "https://www.youtube.com/watch?v=" + str(youtube["data-id"])

            if picture == None:
                p = ""
            else: 
                p = picture["href"]

            if score == None:
                score = ""
            else: 
                score = score["data-percent"]

            content_lis = soup.find_all('li', attrs={'class': "profile"})
            crew = []
            for li in content_lis:
                crew.append(li.getText().split('\n')[1:-1])

            budget = str(soup.select('.facts > p:nth-child(8)'))
            start = budget.find('</strong>') + 10
            end = budget.find('</p>', start)
            budget = budget[start:end]

            rev = str(soup.select('.facts > p:nth-child(9)'))
            start = rev.find('</strong>') + 10
            end = rev.find('</p>', start)
            rev = rev[start:end]

            runtime = str(soup.select('.facts > p:nth-child(7)'))
            start = runtime.find('</strong>') + 10
            end = runtime.find('</p>', start)
            runtime = runtime[start:end]

            lang = str(soup.select('.facts > p:nth-child(6)'))
            start = lang.find('</strong>') + 10
            end = lang.find('</p>', start)
            lang = lang[start:end]
            keywords = soup.select('.keywords > ul:nth-child(2)')

            if keywords == [] :
                keywords = ""
            else: 
                keywords = keywords[0].text[1::][:-1].replace("\n", ",").split(",")
            
            d = str(df.loc[i, "movieId"]) + "|" + str(overview.text).replace('\n','') + "|" +  str(y) + "|" +  str(p) + "|" +  str(crew) + "|" +  str(budget) + "|" +  str(rev) + "|" +  str(runtime) + "|" + str(lang) + "|" + str(keywords) + "\n"
            f.writelines(d)
            display.clear_output(wait=True)
            display.display((i/len(df))*100)
```

```python
data = pd.read_csv('tmdbinfofnl.txt', sep="|", header=None)
data.columns = ["id", "overview", "ytlink", "poster", 'featcrew', 'budget', 'revenue', 'runtime', 'lang', 'keywords']
```

```python
data.info()
```

```python
data2 = pd.read_csv('ml-latest-small/movies.csv')
```

```python
fulldf = pd.merge(data2, data, left_on='movieId', right_on='id', how='left').drop('id', axis=1)
```

```python
fulldf['genres'] = fulldf['genres'].str.split('|')
```

```python
fulldf['keywords'] = fulldf['keywords'].str.replace("'",'')
```

```python
fulldf['year'] = fulldf['title'].str[-6:]
fulldf['title'] = fulldf['title'].str[:-6]
```

```python
fulldf['year'] = fulldf['year'].str[1:-1]
```

```python
keys = range(len(fulldf))
for i in keys:
    if type(fulldf['featcrew'][i]) == float or fulldf['featcrew'][i] == '[]':
        fulldf['featcrew'][i] = ""
    else:
        a = list(eval(fulldf['featcrew'][i].replace("[",'').replace("]", "")))
        names = a[::2]
        roles = a[1::2]
        fulldf['featcrew'][i] = dict(zip(names, roles))
```

```python
a = fulldf['budget'].str.contains('h', regex=True, na=False)
b = fulldf[a]
b['lang'] = b['runtime']
b['runtime'] = b['budget']
b['budget'] = "-"
```

```python
fulldf.update(b)
```

```python
a = fulldf[fulldf['runtime'].str.contains('h', regex=True, na=False) == False]
a['lang'] =  a['runtime']
a["runtime"] = ""
```

```python
fulldf.update(a)
```

```python
fulldf.fillna("-", inplace=True)
```

```python
fulldf[fulldf['ytlink'] == "-"].count()
```

```python
import urllib.request
from bs4 import BeautifulSoup
a = list(fulldf.loc[fulldf['ytlink'] == "-", 'title'])
for i in range(len(a)):
    textToSearch = a[i]
    query = urllib.parse.quote(textToSearch + " trailer")
    url = "https://www.youtube.com/results?search_query=" + query
    response = urllib.request.urlopen(url)
    html = response.read()
    soup = BeautifulSoup(html, 'html.parser')
    for vid in soup.findAll(attrs={'class':'yt-uix-tile-link'}):
        fulldf.loc[fulldf['title'] == textToSearch, 'ytlink'] = 'https://www.youtube.com' + vid['href']
        break
    display.clear_output(wait=True)
    display.display((i/len(a))*100)
```
