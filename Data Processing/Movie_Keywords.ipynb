```python
import pandas as pd
from pandas import DataFrame
import warnings
import numpy as np
import ast
warnings.filterwarnings('ignore')
```

```python
df = pd.DataFrame(pd.read_csv('../Data/movieinfo.csv')) 
```

```python
df.head()
```

```python
lst_col = 'keywords' 
```

```python
x = df.assign(**{lst_col:df[lst_col].str.split(',')})
```

```python
df = pd.DataFrame({col:np.repeat(x[col].values, x[lst_col].str.len())for col in x.columns.difference([lst_col])}).assign(**{lst_col:np.concatenate(x[lst_col].values)})[x.columns.tolist()]
```

```python
df = df.drop(df.columns[[2,3, 4, 5, 6, 7, 8, 9, 11]], axis=1)
```

```python
df = df.drop(df.columns[[1]], axis=1)
```

```python
df
```

```python
df['keywords'] = df['keywords'].str.strip('[')
df['keywords'] = df['keywords'].str.strip(']')
df['keywords'] = df['keywords'].str.strip('""')
df['keywords'] = df['keywords'].str.strip('"')
```

```python
df = df[df.keywords != "-"]
```

```python
from pandas.api.types import is_string_dtype
is_string_dtype(df['keywords'])
```

```python
df = df[~df['keywords'].str.contains('"')]
```

```python
df.to_csv('../Data/movie_keys.csv', index=False)
```
